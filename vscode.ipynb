```python
import pandas as pd
import hashlib as hb
```

```python
vscode = pd.read_csv('./data/1092_viscode_logs.csv')
```

```python
vscode['username'] = vscode['username'].str.split('@',expand=True).get(0)
vscode = vscode[vscode['username'].str.isdigit() == True]
vscode = vscode[vscode['username'].apply(lambda x: len(str(x))>=9)]
```

```python
def md5(i):
    hl = hb.md5()
    
    hl.update(str(i).encode(encoding='utf-8'))

    #print('加密前為 ：' + str(i))
    #print('加密後為 ：' + hl.hexdigest())
    
    return hl.hexdigest()

vscode['username'] = vscode['username'].apply(md5)
```

```python
vscode.to_csv('./output/vscode.csv',encoding='utf8') 
```
